**Functions**
A function is a way to encapsulate a certain calculation or algorithm. Use them as often as possible, especially if you find you are repeating yourself. A good acronym to remember is DRY = "Don't Repeat Yourself". It is the opposite of WET = "Write Everything Twice" or "We Enjoy Typing".

```python
def f(x,m,b):
    return m*x + b
```

```python
f(1,2,3)
```

```python
def sum(start, stop):
    s = 0
    for x in range(start,stop+1):
        s = s + x
    return s
```

```python
sum(1,10)
```

**Keyword arguments**
You may encounter (oir even define yourself) functions whose arguments have defaults. These are very useful because the *order* of the arguments does not matter, so you can't accidentally call such a function with the arguments reversed.

```python
def another_sum(start,stop,step=1):
    s = 0
    for x in range(start,stop+1,step):
        s = s + x
    return s    
```

```python
another_sum(1,10) # returns 1 + 2 + ... + 10 = 55
```

```python
another_sum(1,10,2) # returns 1 + 3 + 5 + 7 + 9 = 25
```

**Classes and Objects**

Classes are ways to include data and code that belong together. For example, here is a very simple fraction class.

```python
class Fraction:
    def __init__(self,a,b):
        self.numerator = a
        self.denominator = b
    def to_decimal(self):
        return self.numerator / self.denominator
    def recipricol(self):
        return Fraction(self.denominator, self.numerator)
```

To construct new `Fraction` objects using the Fraction class, we call `Fraction` as though it were a function. And to access the methods in a fraction object, we use the `.` operator.

```python
x = Fraction(2,5)
y = x.recipricol()
y.denominator
```

```python
x.to_decimal()
```

```python
y.to_decimal()
```

You can get a list of all the methods in an object using the `dir()` method, as in:

```python
dir(x)
```

Plotting Functions
---
Plotting functions in Jupyter and Python can be achieved via many different libraries. One of the most common, and the one we will use in this course, is called `matplotlib`, which we need to import before we can use. We also have to tell Jupyter to put the plots we make directly into our notebook. Commands that talk to Jupyter are prefixed by a `%` sign.

More information about the many types of plots you can make can be found [here](https://matplotlib.org/).

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

To plot a function $y=f(x$), we first define a list of $x$ values. This can be done using the `numpy` method called `linspace` as follows.

```python
x = np.linspace(0,1,101) # returns an array of 101 floats, linearly spaced from 0 to 1 inclusive.
```

```python
x[0:10]
```

Next, we need to evaluate the a function on all of these points to define a list of `y` values. Magically, `numpy` allows you to apply methods like `sin`, `cos` as well as operators like `+` and `*` on `linspace` objects as well as individual values.

```python
y = np.sin(x)
y[0:10]
```

```python
z = np.sin(x) / (1 + x)
z[0:10]
```

Finally, you can plot the function with the `plot` method of `pyplot`.

```python
plt.plot(x,y);
```

We can also put multiple curves on top of each other and provide a legend.

```python
t = np.linspace(-4*np.pi,4*np.pi,100);
plt.plot(t,np.sin(t) / (1+t*t),label="function 1");
plt.plot(t,np.cos(t) / (1+t*t),label="function 2");
plt.legend();
```

**User defined functions**

You can also define and plot your own functions to plot.

```python
def f(x):
    return x*x + 2*x + 1
```

```python
x = np.linspace(0,1,101)
plt.plot(x,f(x),label="f(x)");
plt.plot(x,f(f(x)),label="f(f(x))");
plt.legend();
```

**Multiple Plots**
You can plot mutiple plots in a grid. You can also add specifications to how the curves are drawn.

```python
fig, ax = plt.subplots(2,2);
t = np.linspace(0,2*np.pi,20);
ax[0][0].plot(t,np.sin(t),'r',label="function 0,0");
ax[0][1].plot(t,np.cos(t),'g-',label="function 0,1");
ax[1][0].plot(t,f(t)/40,'b-.',label="function 1,0");
ax[1][1].plot(t,np.exp(t)/100,'yo',label="function 1,1");
ax[0][0].legend();
ax[0][1].legend();
ax[1][0].legend();
ax[1][1].legend();
```

Parametric Plots
---
In a parametric plot, we have one independent variable, say time `t`, and two dependent variables, say `x(t)` and `y(t)`. These two functions of `t` can be ploted in the `x`-`y` plane.

```python
t = np.linspace(0,50,400);
x = t*np.sin(t);
y = t*np.cos(t);
plt.plot(x,y);
```
